```python
import numpy as np
import matplotlib.pyplot as plt
from math import tanh
```

```python
data1 = np.arange(-10, 10)

z = [0 if i < 0 else 1 for i in data1]

plt.plot(data1, z)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Unit Step")
plt.show()
```

```python
data2 = np.arange(-20, 20)

def signum(data):
  if data < 0:
    return -1
  if data > 0:
    return 1
  return 0

z = [signum(i) for i in data2]
# len(data2)

plt.plot(data2, z)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Sign (Signum)")
plt.show()
```

```python
data3 = np.arange(-5, 5)

linear = lambda data, slope, intercept: (data * slope) + intercept

z = [linear(i, 10, 1) for i in data3]

plt.plot(data3, z)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Linear")
plt.show()
```

```python
data4 = np.arange(-15, 15)

def piecewise_linear(data, min):
  min = min + 5 if min < 0 else min - 5

  if data < min:
    return (data * 9) + 2
  if data >= min and data < 0:
    return data + 1
  if data == 0:
    return 0
  if data > 0:
    return (data + 3) * 5.2
  
z = [piecewise_linear(i, -15) for i in data4]

plt.plot(data4, z)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Piecewise Linear")
plt.show()
```

```python
data5 = np.arange(-8, 8)

sigmoid = lambda x: 1 / (1 + np.exp(-x))

z = [sigmoid(i) for i in data5]

plt.plot(data5, z)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Sigmoid")
plt.show()
```

```python
data6 = np.arange(-12, 12)
z = [tanh(i) for i in data6]

plt.plot(data6, z)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Tanh")
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data_iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data = data_iris["sepal width (cm)"]
data = list(data)
data.sort()
# data
```

```python
z = [0 if i < 0 else 1 for i in data]

plt.plot(data, z)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("f(X)")
plt.title("Unit Step")
plt.show()
```

```python
z = [signum(i) for i in data]

plt.plot(data, z)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("f(X)")
plt.title("Signum")
plt.show()
```

```python
z = [linear(i, 10, 3) for i in data]

plt.plot(data, z)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("f(X)")
plt.title("Linear")
plt.show()
```

```python
def piecewise_linear2(data, min):
  if data == min:
    return data
  if data > min and data < min + 1.2:
    return (data * 8) + 10
  return ((data + 1) * 9.2) / 1.2

z = [piecewise_linear2(i, min(data)) for i in data]

plt.plot(data, z)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("f(X)")
plt.title("Piecewise Linear")
plt.show()
```

```python
z = [sigmoid(i) for i in data]

plt.plot(data, z)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("f(X)")
plt.title("Sigmoid")
plt.show()
```

```python
z = [tanh(i) for i in data]

plt.plot(data, z)
plt.xlabel("Sepal Width (cm)")
plt.ylabel("f(X)")
plt.title("Tanh")
plt.show()
```

# OBSERVATION
Signum & Unit step functions only work for data set with both -ve and +ve values, otherwise it just displays a single line.
