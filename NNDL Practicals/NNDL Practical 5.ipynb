Explore the commands relevant to the following identities/outputs with different input
values.

1. Identity
2. ones
3. zeros
4. sum
5. size
6. Linear spaced data
7. random

```python
import numpy as np
```

```python
def printn(*args):
  print(*args, end="\n\n")
```

```python
printn(np.identity(3))
print(np.identity(10))
```

```python
printn(np.ones(5))
printn(np.ones((2, 3)))
print(np.ones((3, 4, 2)))
```

```python
printn(np.zeros(5))
printn(np.zeros((2, 3)))
print(np.zeros((3, 4, 2)))
```

```python
printn(np.sum([2, 3, 4]))
printn(np.sum([[0, 1, 2], [3, 4, 5]]))
printn(np.sum([[0, 1, 2], [3, 4, 5]], axis=0))
printn(np.sum([[0, 1, 2], [3, 4, 5]], axis=1))
print(np.sum([2, 3, 4, 5], dtype=np.float32))
```

```python
printn(np.size(3))
printn(np.size([3, 1]))
printn(np.size([[3, 4, 2], [2, 3, 8]]))
printn(np.size([[3, 4, 2], [2, 3, 8]], axis=0))
print(np.size([[3, 4, 2], [2, 3, 8]], axis=1))
```

```python
printn(np.linspace(2, 5))
printn(np.linspace(2, 5, num=10))
printn(np.linspace(2, 5, num=10, retstep=True))
printn(np.linspace(2, 5, num=10, endpoint=False))
```

```python
printn(np.random.rand(2))
printn(np.random.rand(2, 3))
printn(np.random.randint(4))
printn(np.random.randint(4, 10))
printn(np.random.randint(4, 10, size=10))
printn(np.random.randint(4, 10, size=(3, 2)))
printn(np.random.randint([2, 3, 5], 10))
```
