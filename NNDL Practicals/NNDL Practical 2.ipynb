```python
import numpy as np
```

```python
data = [
    [0, 0], [0, 1], [1, 0], [1, 1]
]

w1, w2, w0 = 1.5, .5, -.5
W = [[w1], [w2]]
```

```python
def apply_eqn(data, weights, bias):
  return np.dot(data, weights)[0] + bias
```

```python
def work(data, operator, weights, bias):
  print(f"For { operator.upper() if operator != '!=' else 'XOR' }:")
  error = 0

  for i in data:
    equated_op = apply_eqn(i, weights, bias)
    real_op = eval(f"i[0] { operator } i[1]")
    print(f"Input: { i } Equated output: { equated_op } Real output: { 1 if real_op else 0 }")
    error += equated_op - real_op
  print(f"Error: {error}")
```

```python
work(data, "and", W, w0)
print()
work(data, "or", W, w0)
print()
work(data, "!=", W, w0)
```
