```python
X = [
    [1, 1],
     [1, -1],
     [-1, 1],
     [-1, -1]
]

labels = [-1, 1, 1, -1]
v1 = v2 = b3 = 0.3
threshold = 0
alpha = 0.1
epochs = 2
```

```python
def fun(X):
  return 1 if X >= threshold else -1
```

```python
w11 = 0.3; w21 = 0.1; b1 = 0.2
w12 = 0.2; w22 = 0.3; b2 = 0.1
# w11 = w12 = w21 = w22 = b1 = b2 = 0.1

for epoch in range(epochs):
  print(f"For epoch { epoch + 1 }:")
  for i in range(len(X)):
    x1, x2 = X[i]
    target = labels[i]
    print(f"S({ x1 }, { x2 }): { target }")

    z1 = round((w11 * x1) + (w21 * x2) + b1, 4)
    z2 = round((w12 * x1) + (w22 * x2) + b2, 4)
    print(f"z1 = { z1 }, z2 = { z2 }")

    fz1 = fun(z1); fz2 = fun(z2)
    Y = round((v1 * fz1) + (v2 * fz2) + b3, 4)
    print(f"Y = { Y }")
    error_z1 = round(target - z1, 4); error_z2 = round(target - z2, 4)

    if fun(Y) == target:
      print("f(Y) == target; no weight updation required.\n")
      continue
    if target == 1:
      closest_to_0 = min(abs(z1), abs(z2))

      if abs(z1) == closest_to_0:
        w11 = round(w11 + ((alpha * error_z1) * x1), 4)
        w21 = round(w21 + ((alpha * error_z1) * x2), 4)
        b1 = round(b1 + (alpha * error_z1), 4)
        print(f"w11n = { w11 }, w21n = { w21 }, b1n = { b1 }")
      if abs(z2) == closest_to_0:
        w12 = round(w12 + ((alpha * error_z2) * x1), 4)
        w22 = round(w22 + ((alpha * error_z2) * x2), 4)
        b2 = round(b2 + (alpha * error_z2), 4)
        print(f"w12n = { w12 }, w22n = { w22 }, b2n = { b2 }")
    else:
      if z1 >= 0:
        w11 = round(w11 + ((alpha * error_z1) * x1), 4)
        w21 = round(w21 + ((alpha * error_z1) * x2), 4)
        b1 = round(b1 + (alpha * error_z1), 4)
        print(f"w11n = { w11 }, w21n = { w21 }, b1n = { b1 }")
      if z2 >= 0:
        w12 = round(w12 + ((alpha * error_z2) * x1), 4)
        w22 = round(w22 + ((alpha * error_z2) * x2), 4)
        b2 = round(b2 + (alpha * error_z2), 4)
        print(f"w12n = { w12 }, w22n = { w22 }, b2n = { b2 }")
    print()
```
